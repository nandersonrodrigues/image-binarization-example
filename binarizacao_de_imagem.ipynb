```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
#cv2.imshow() causa problemas no google collab, usar cv2_imshow 
from google.colab.patches import cv2_imshow 
```

```python
#monaliza image
img = Image.open("monaliza.jpeg")
img
```

##Converte RGB para escala de cinza

```python
img = cv.imread("monaliza.jpeg")
```

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

## Binarizao da imagem

###aplicando o blur

```python
img_blur = cv.medianBlur(img_gray, 5)
cv2_imshow(img_blur)
```

###Thresholding (limiar) adaptativo

####Usando mtodo Gausiano

```python
img_thresholding = cv.adaptiveThreshold(img_blur, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY_INV,7,2)

cv2_imshow(img_thresholding)
```

####Usando mtodo da mdia

```python
img_thresholding = cv.adaptiveThreshold(img_blur, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY_INV,5,2)

cv2_imshow(img_thresholding)
```
